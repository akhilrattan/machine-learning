```python
from pathlib import Path

DATASET = Path("/content/trashnetraw")

print(DATASET.exists())
print((DATASET / "train/images").exists())
print((DATASET / "valid/images").exists())
print((DATASET / "test/images").exists())
```

```python
import yaml

yaml_path = "/content/trashnetraw/data.yaml"

with open(yaml_path, "r") as f:
    data = yaml.safe_load(f)

data["path"] = "/content/processed"

with open(yaml_path, "w") as f:
    yaml.dump(data, f, sort_keys=False)

print(data)
```

```python
from ultralytics import YOLO

model = YOLO("yolov8s.pt")

results = model.train(
    data="/content/trashnetraw/data.yaml",
    epochs=30,
    imgsz=640,
    batch=16,
    workers=2,
    device=0,
    project="/content/trashnet_yolo_results",
    name="final_training",
    pretrained=True,
    plots=True,
    save=True,
    save_period=5,
    patience=5,
    exist_ok=True
)
```
